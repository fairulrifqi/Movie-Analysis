# Portofolio 5: Machine Learning Overflow in Movie Analysis

```python
# Convert 'popularity' column to numeric, handling errors
new_data['popularity'] = pd.to_numeric(new_data['popularity'], errors='coerce').astype(int)
new_data['budget'] = pd.to_numeric(new_data['budget'], errors='coerce').astype(int)
new_data['runtime'] = pd.to_numeric(new_data['runtime'], errors='coerce').astype(int)
new_data['rating'] = pd.to_numeric(new_data['rating'], errors='coerce')
```

```python
new_data.select_dtypes(include='number')
```

```python
def select_columns(df: pd.DataFrame, features: list, target: list = 'rating'):
  x = df[features]
  y = (df[target] >= 3).astype(int)
  return x, y
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = ['budget', 'popularity', 'vote_average']
x, y = select_columns(new_data, features = features)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(x_train)
```

```python
x_train_scaled = minmax.transform(x_train)
x_test_scaled = minmax.transform(x_test)
```

```python
def scaling(df, scaler):
  scaled_array = scaler.transform(df)
  scaled_df = pd.DataFrame(scaled_array, columns=scaler.get_feature_names_out())
  return scaled_df
```

```python
scaled_x_train = scaling(x_train, minmax)
scaled_x_test = scaling(x_test, minmax)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
dtc = DecisionTreeClassifier()
```

```python
knn.fit(scaled_x_train, y_train)
dtc.fit(scaled_x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Assuming 'knn_new' is your KNeighborsClassifier instance
# and you have already performed a GridSearchCV
# For demonstration, let's assume 'knn_new' has been fitted

# Create GridSearchCV object
# Replace parameters with your desired search space
param_grid = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
knn_best = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)

# Fit the GridSearchCV object to find the best model
knn_best.fit(x_train, y_train)  # Assuming x_train and y_train are defined

# Now, 'knn_best' holds the best model found by GridSearchCV
# You can use it for prediction
knn_best.predict(x_test)
```

```python
recall_ = []
precision_ = []
accuracy_ = []
f1_ = []

for i in range(1, 11):
  y_best_knn_prob = knn_best.predict_proba(x_test)[:, 1]
  recall_.append(recall_score(y_test, (y_best_knn_prob >= i/10)))
  precision_.append(precision_score(y_test, (y_best_knn_prob >= i/10)))
  accuracy_.append(accuracy_score(y_test, (y_best_knn_prob >= i/10)))
  f1_.append(f1_score(y_test, (y_best_knn_prob >= i/10)))
```

```python
metrics_ = pd.DataFrame({'recall': recall_,
              'prescision': precision_,
              'accuracy': accuracy_,
              'f1_score': f1_})
```

```python
metrics_['fpr'] = 1 - metrics_['recall']
```

```python
metrics_.plot(y='recall', x='fpr')
```
